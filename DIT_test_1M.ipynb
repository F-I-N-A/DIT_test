```python
# Импортируем библиотеку
import pandas as pd
```

```python
# Загружаем данные в датафрейм
data = pd.read_csv('data_for_test_task.csv', sep=';')
data.head()
```

```python
# Создаем результиирующий датафрейм 
df = data.pivot(index=['DOCUMENT_ID', 'LINE_ID'], columns='COL')
df = df['VAL'].reset_index()
df.columns.name = None
df.head()
```

```python
# Дополняем значениями столбец 'ид_пациента'
df['ид_пациента'] = df['DOCUMENT_ID']
df.head()
```
