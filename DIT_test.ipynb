```python
# Импортируем библиотеку
import pandas as pd
```

```python
# Загружаем данные в датафрейм
data = pd.read_csv('data.csv', sep=';')
data.head()
```

```python
# Формируем список с названиями столбцов для результирующего датафрейма
cols = data['COL'].unique().tolist()
columns = ['DOCUMENT_ID', 'LINE_ID'] + cols
columns
```

```python
# Определим количество строк в результирующем датафрейме, исходя из уникальных сочетаний полей 'DOCUMENT_ID' и 'LINE_ID'
df1 = data[['DOCUMENT_ID', 'LINE_ID']]
df1 = df1.drop_duplicates()

len(df1)
```

```python
# Создаем словарь, на основе которого будет создан итоговый датафрейм
dict = {}

# Добавляем значения для столбцов 'DOCUMENT_ID' и 'LINE_ID'
dict['DOCUMENT_ID'] = df1['DOCUMENT_ID'].to_list()
dict['LINE_ID'] = df1['LINE_ID'].to_list()
```

```python
# Создадим пустые списки заданной длины в качестве значений словаря для остальных столбцов
for c in cols:
    dict[c] = [''] * len(df1)
```

```python
# Добавляем значения для столбца 'ид_пациента' (совпадает со значениями стобца'DOCUMENT_ID')
dict['ид_пациента'] = dict['DOCUMENT_ID']
```

```python
# Заполняем оставшиеся значения в словарь
r = 0
for i in range(data['DOCUMENT_ID'].max()+1):
    
    for j in range(1, data[data['DOCUMENT_ID']==i]['LINE_ID'].unique().max()+1):
        
        tmp_df = data[(data['DOCUMENT_ID']==i) & (data['LINE_ID']==j)]
        tmp_df = tmp_df[['COL', 'VAL']].set_index('COL')
        
        for elem in tmp_df.index:
            dict[elem][r] = tmp_df.loc[elem, 'VAL']
        r += 1
      
dict
```

```python
# Создаем результиирующий датафрейм на основе сформированного словаря со значениями
df = pd.DataFrame(dict)
df
```
